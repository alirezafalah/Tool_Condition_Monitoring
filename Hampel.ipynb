```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from hampel import hampel

df_intact = pd.read_csv(r"data\drill_intact.csv") 

# Scaling
df_intact["Sum of Pixels"] = (df_intact["Sum of Pixels"] - df_intact["Sum of Pixels"].min()) / (
    df_intact["Sum of Pixels"].max() - df_intact["Sum of Pixels"].min()
)

# Shifting
min_index = df_intact["Sum of Pixels"].idxmin()
df_intact_shifted = pd.concat([df_intact.loc[min_index:], df_intact.loc[:min_index - 1]])
df_intact_shifted = df_intact_shifted.reset_index(drop=True)
df_intact_shifted["Degree"] = df_intact_shifted["Degree"] - df_intact_shifted.iloc[0]["Degree"]
df_intact_shifted.loc[df_intact_shifted["Degree"] < 0, "Degree"] += 360

# Plot
plt.figure(figsize=(8, 6))
plt.plot(df_intact_shifted["Degree"], df_intact_shifted["Sum of Pixels"], marker="o", linestyle="-")
plt.xlabel("Degree")
plt.ylabel("Sum of Pixels")
plt.title("Processed")
plt.grid(True)
plt.show()
```

```python
#HAMPLE 
df_intact_shifted_original = df_intact_shifted.copy()

result = hampel(df_intact_shifted_original["Sum of Pixels"].values, window_size=65, n_sigma=2.0)
df_intact_shifted_filtered = df_intact_shifted_original.copy()
df_intact_shifted_filtered["Sum of Pixels"] = result.filtered_data

plt.figure(figsize=(8, 6))
plt.plot(df_intact_shifted["Degree"], df_intact_shifted["Sum of Pixels"], marker="o", linestyle="-")
plt.xlabel("Degree")
plt.ylabel("Sum of Pixels")
plt.title("Hample (n_sigma=2.0, window_szie=65)")
plt.grid(True)
plt.show()
```
